```python
# ============================================
# SISTEM PREDIKSI OBESITAS - RANDOM FOREST
# Training Model Lengkap dengan Visualisasi
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (classification_report, confusion_matrix, 
                             accuracy_score, precision_score, recall_score, 
                             f1_score, roc_auc_score, roc_curve)
import joblib
import warnings
warnings.filterwarnings('ignore')

# Set style untuk visualisasi
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("=" * 60)
print("SISTEM PREDIKSI OBESITAS - RANDOM FOREST")
print("=" * 60)

# ============================================
# 1. LOAD DATA
# ============================================
print("\n[1] LOADING DATASET...")
df = pd.read_csv('data/raw/ObesityDataSet_raw_and_data_sinthetic.csv')

print(f"✓ Dataset berhasil dimuat!")
print(f"  Jumlah baris: {df.shape[0]}")
print(f"  Jumlah kolom: {df.shape[1]}")

# Tampilkan info dataset
print("\n[INFO] Struktur Dataset:")
print(df.info())

print("\n[INFO] 5 Baris Pertama:")
print(df.head())

print("\n[INFO] Statistik Deskriptif:")
print(df.describe())

# ============================================
# 2. EXPLORATORY DATA ANALYSIS
# ============================================
print("\n[2] EXPLORATORY DATA ANALYSIS...")

# Cek missing values
print("\n[CHECK] Missing Values:")
missing = df.isnull().sum()
print(missing[missing > 0] if missing.sum() > 0 else "✓ Tidak ada missing values!")

# Distribusi kelas target
print("\n[INFO] Distribusi Kelas Target:")
print(df['NObeyesdad'].value_counts())

# Visualisasi 1: Distribusi Kelas
plt.figure(figsize=(12, 6))
df['NObeyesdad'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Distribusi Kategori Obesitas', fontsize=16, fontweight='bold')
plt.xlabel('Kategori Obesitas', fontsize=12)
plt.ylabel('Jumlah Sampel', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('results/figures/01_class_distribution.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ Visualisasi disimpan: results/figures/01_class_distribution.png")

# Visualisasi 2: Correlation Heatmap
plt.figure(figsize=(14, 10))
numeric_cols = df.select_dtypes(include=[np.number]).columns
corr_matrix = df[numeric_cols].corr()
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            square=True, linewidths=0.5, cbar_kws={"shrink": 0.8})
plt.title('Correlation Matrix - Variabel Numerik', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('results/figures/02_correlation_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ Visualisasi disimpan: results/figures/02_correlation_heatmap.png")

# ============================================
# 3. PREPROCESSING
# ============================================
print("\n[3] PREPROCESSING DATA...")

# Pisahkan fitur dan target
X = df.drop('NObeyesdad', axis=1)
y = df['NObeyesdad']

print(f"✓ Fitur (X): {X.shape}")
print(f"✓ Target (y): {y.shape}")

# Encode categorical variables
label_encoders = {}
categorical_cols = X.select_dtypes(include=['object']).columns

print(f"\n[ENCODING] Variabel Kategorik: {len(categorical_cols)} kolom")
for col in categorical_cols:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])
    label_encoders[col] = le
    print(f"  ✓ {col}: {len(le.classes_)} kategori")

# Encode target variable
le_target = LabelEncoder()
y_encoded = le_target.fit_transform(y)
print(f"\n[ENCODING] Target Variable: {len(le_target.classes_)} kelas")
for i, class_name in enumerate(le_target.classes_):
    print(f"  {i}: {class_name}")

# Split data
print("\n[SPLIT] Membagi data menjadi Training & Testing Set...")
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)
print(f"✓ Training set: {X_train.shape[0]} sampel")
print(f"✓ Testing set: {X_test.shape[0]} sampel")

# Feature Scaling
print("\n[SCALING] Normalisasi fitur numerik...")
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
print("✓ Scaling selesai!")

# ============================================
# 4. TRAINING MODEL (Baseline)
# ============================================
print("\n[4] TRAINING BASELINE MODEL...")
print("=" * 60)

rf_baseline = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)

print("Training baseline Random Forest...")
rf_baseline.fit(X_train_scaled, y_train)
print("✓ Training selesai!")

# Prediksi
y_pred_baseline = rf_baseline.predict(X_test_scaled)

# Evaluasi baseline
acc_baseline = accuracy_score(y_test, y_pred_baseline)
print(f"\n[BASELINE] Accuracy: {acc_baseline:.4f} ({acc_baseline*100:.2f}%)")

# ============================================
# 5. HYPERPARAMETER TUNING (Grid Search)
# ============================================
print("\n[5] HYPERPARAMETER OPTIMIZATION...")
print("=" * 60)

param_grid = {
    'n_estimators': [100, 150, 200],
    'max_depth': [10, 15, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2'],
    'criterion': ['gini', 'entropy']
}

print("Parameter Grid:")
for param, values in param_grid.items():
    print(f"  {param}: {values}")

total_combinations = np.prod([len(v) for v in param_grid.values()])
print(f"\nTotal kombinasi: {total_combinations}")
print("Memulai Grid Search (10-fold CV)...")
print("⏳ Ini akan memakan waktu beberapa menit...\n")

rf = RandomForestClassifier(random_state=42, n_jobs=-1)
grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=10,
    scoring='accuracy',
    n_jobs=-1,
    verbose=2
)

grid_search.fit(X_train_scaled, y_train)

print("\n✓ Grid Search selesai!")
print(f"\n[BEST] Best CV Score: {grid_search.best_score_:.4f}")
print(f"[BEST] Best Parameters:")
for param, value in grid_search.best_params_.items():
    print(f"  {param}: {value}")

# Simpan hasil Grid Search
results_df = pd.DataFrame(grid_search.cv_results_)
results_df = results_df.sort_values('rank_test_score')
results_df[['params', 'mean_test_score', 'std_test_score', 'rank_test_score']].head(10).to_csv(
    'results/grid_search_top10.csv', index=False
)
print("\n✓ Top 10 hasil Grid Search disimpan: results/grid_search_top10.csv")

# ============================================
# 6. FINAL MODEL
# ============================================
print("\n[6] TRAINING FINAL MODEL...")
print("=" * 60)

best_model = grid_search.best_estimator_
print("✓ Model terbaik dari Grid Search dipilih")

# Cross-validation score
cv_scores = cross_val_score(best_model, X_train_scaled, y_train, cv=10, scoring='accuracy')
print(f"\n[CV] Cross-Validation Scores (10-fold):")
for i, score in enumerate(cv_scores, 1):
    print(f"  Fold {i:2d}: {score:.4f}")
print(f"  Mean: {cv_scores.mean():.4f} ± {cv_scores.std():.4f}")

# ============================================
# 7. EVALUASI MODEL
# ============================================
print("\n[7] EVALUASI MODEL FINAL...")
print("=" * 60)

# Prediksi
y_pred = best_model.predict(X_test_scaled)
y_pred_proba = best_model.predict_proba(X_test_scaled)

# Metrik evaluasi
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print("\n[METRICS] Performa Model pada Testing Set:")
print(f"  Accuracy:  {accuracy:.4f} ({accuracy*100:.2f}%)")
print(f"  Precision: {precision:.4f}")
print(f"  Recall:    {recall:.4f}")
print(f"  F1-Score:  {f1:.4f}")

# Classification Report
print("\n[REPORT] Classification Report Detail:")
print(classification_report(y_test, y_pred, 
                           target_names=le_target.classes_,
                           digits=4))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(12, 10))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=le_target.classes_,
            yticklabels=le_target.classes_,
            cbar_kws={'label': 'Jumlah Sampel'})
plt.title('Confusion Matrix - Random Forest', fontsize=16, fontweight='bold')
plt.ylabel('Kelas Aktual', fontsize=12)
plt.xlabel('Kelas Prediksi', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig('results/figures/03_confusion_matrix.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ Confusion Matrix disimpan: results/figures/03_confusion_matrix.png")

# Feature Importance
feature_importance = pd.DataFrame({
    'feature': X.columns,
    'importance': best_model.feature_importances_
}).sort_values('importance', ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(data=feature_importance, x='importance', y='feature', palette='viridis')
plt.title('Feature Importance - Random Forest', fontsize=16, fontweight='bold')
plt.xlabel('Importance Score', fontsize=12)
plt.ylabel('Features', fontsize=12)
plt.tight_layout()
plt.savefig('results/figures/04_feature_importance.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ Feature Importance disimpan: results/figures/04_feature_importance.png")

print("\n[INFO] Top 10 Fitur Terpenting:")
print(feature_importance.head(10))

# ROC Curve (One-vs-Rest untuk multi-class)
from sklearn.preprocessing import label_binarize
y_test_bin = label_binarize(y_test, classes=range(len(le_target.classes_)))
n_classes = y_test_bin.shape[1]

plt.figure(figsize=(12, 8))
for i in range(n_classes):
    fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_pred_proba[:, i])
    auc_score = roc_auc_score(y_test_bin[:, i], y_pred_proba[:, i])
    plt.plot(fpr, tpr, lw=2, 
             label=f'{le_target.classes_[i]} (AUC = {auc_score:.3f})')

plt.plot([0, 1], [0, 1], 'k--', lw=2, label='Random Guess')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate', fontsize=12)
plt.ylabel('True Positive Rate', fontsize=12)
plt.title('ROC Curves - Multi-class Classification', fontsize=16, fontweight='bold')
plt.legend(loc="lower right", fontsize=9)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('results/figures/05_roc_curves.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ ROC Curves disimpan: results/figures/05_roc_curves.png")

# Model Comparison
comparison_df = pd.DataFrame({
    'Model': ['Baseline RF', 'Optimized RF'],
    'Accuracy': [acc_baseline, accuracy],
    'Precision': [
        precision_score(y_test, y_pred_baseline, average='weighted'),
        precision
    ],
    'Recall': [
        recall_score(y_test, y_pred_baseline, average='weighted'),
        recall
    ],
    'F1-Score': [
        f1_score(y_test, y_pred_baseline, average='weighted'),
        f1
    ]
})

print("\n[COMPARISON] Baseline vs Optimized Model:")
print(comparison_df.to_string(index=False))

# ============================================
# 8. SAVE MODEL
# ============================================
print("\n[8] MENYIMPAN MODEL...")
print("=" * 60)

joblib.dump(best_model, 'models/obesity_model.pkl')
print("✓ Model disimpan: models/obesity_model.pkl")

joblib.dump(scaler, 'models/scaler.pkl')
print("✓ Scaler disimpan: models/scaler.pkl")

joblib.dump(label_encoders, 'models/label_encoders.pkl')
print("✓ Label Encoders disimpan: models/label_encoders.pkl")

joblib.dump(le_target, 'models/target_encoder.pkl')
print("✓ Target Encoder disimpan: models/target_encoder.pkl")

# Save metadata
metadata = {
    'features': list(X.columns),
    'classes': list(le_target.classes_),
    'best_params': grid_search.best_params_,
    'accuracy': accuracy,
    'precision': precision,
    'recall': recall,
    'f1_score': f1,
    'cv_mean': cv_scores.mean(),
    'cv_std': cv_scores.std()
}

import json
with open('models/model_metadata.json', 'w') as f:
    json.dump(metadata, f, indent=4)
print("✓ Metadata disimpan: models/model_metadata.json")

# ============================================
# SELESAI
# ============================================
print("\n" + "=" * 60)
print("✓ TRAINING SELESAI!")
print("=" * 60)
print("\nFile yang dihasilkan:")
print("1. models/obesity_model.pkl - Model Random Forest")
print("2. models/scaler.pkl - StandardScaler")
print("3. models/label_encoders.pkl - Label Encoders")
print("4. models/target_encoder.pkl - Target Encoder")
print("5. models/model_metadata.json - Model Metadata")
print("6. results/figures/*.png - Visualisasi")
print("7. results/grid_search_top10.csv - Hasil Grid Search")
print("\nSiap untuk deployment! 🚀")
```
